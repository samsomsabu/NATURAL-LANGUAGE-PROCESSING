```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer

# Sample document
document = """
Lamborghini, renowned for its luxurious and high-performance sports cars, has a fascinating history dating back to the mid-20th century.

Ferruccio Lamborghini, the founder of Lamborghini Automobili, initially established his company not in the automobile industry, but rather in the business of manufacturing tractors. In the aftermath of World War II, Lamborghini began producing agricultural machinery from surplus military hardware.

Ferruccio Lamborghini had a profound passion for luxury cars and owned several high-end vehicles, including Ferraris. However, he was dissatisfied with the performance and reliability of his Ferraris. Legend has it that Lamborghini's frustration with Ferrari's quality led him to visit Enzo Ferrari himself to discuss the issues. However, Enzo Ferrari allegedly dismissed Lamborghini's complaints, stating that "the problem is not with the car but with the driver." This encounter reportedly motivated Lamborghini to start his own car manufacturing company, aiming to create superior sports cars that would rival Ferrari's.

In 1963, Lamborghini Automobili was officially established in Sant'Agata Bolognese, Italy. The company's first production car, the Lamborghini 350 GT, debuted in 1964 at the Geneva Motor Show. It featured a V12 engine, a sleek design by renowned designer Franco Scaglione, and luxurious amenities, marking the beginning of Lamborghini's legacy in the automotive world.

Throughout the 1960s and 1970s, Lamborghini continued to innovate and produce iconic sports cars that captured the attention of enthusiasts worldwide. Notable models from this era include the Miura, often considered the world's first supercar, the stunning Countach, and the elegant Espada.

However, financial difficulties plagued Lamborghini in the 1970s, leading to changes in ownership and management. Despite these challenges, the brand endured and experienced a resurgence in the 1980s with models like the Jalpa and the iconic Lamborghini Countach LP5000 Quattrovalvole.

The 1990s saw Lamborghini under the ownership of various companies, including Chrysler and later Audi AG, a subsidiary of Volkswagen Group. Under Audi's stewardship, Lamborghini experienced significant advancements in technology and design, resulting in groundbreaking models such as the Diablo, Murciélago, and Gallardo.

In the 21st century, Lamborghini continued to push the boundaries of automotive engineering with the introduction of innovative models like the Aventador and the Huracán. Additionally, the company expanded its offerings with limited-edition hypercars like the Veneno, Centenario, and Sian.

Today, Lamborghini remains an icon of luxury, performance, and Italian automotive craftsmanship, continuing to captivate enthusiasts with its cutting-edge designs, powerful engines, and uncompromising performance.
"""

# Preprocessing
def preprocess_text(text):
    sentences = sent_tokenize(text)
    word_tokens = [word_tokenize(sentence) for sentence in sentences]
    stop_words = set(stopwords.words('english'))
    filtered_sentences = [[word for word in words if word.lower() not in stop_words] for words in word_tokens]
    stemmer = PorterStemmer()
    stemmed_sentences = [[stemmer.stem(word) for word in words] for words in filtered_sentences]
    return stemmed_sentences

# Answer extraction
def extract_answers(question, text):
    preprocessed_text = preprocess_text(text)
    for sentence in preprocessed_text:
        if question.lower() in ' '.join(sentence).lower():
            return ' '.join(sentence)

# Accept user input for the question
user_question = input("Ask a question: ")

# Answer the user's question
answer = extract_answers(user_question, document)
print("Question:", user_question)
print("Answer:", answer)
```
